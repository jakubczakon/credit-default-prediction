Let's take a quick look at all the tables

```python
for filename in [f for f in os.listdir(DIR) if f.endswith('.zip')]:
    display('FILENAME: {}'.format(filename))
    df = pd.read_csv(os.path.join(DIR, filename), nrows=NROWS)
    display(df.head())
```

We can see that tables are connected on the **SK_ID_CURR** unique id. 

Let's see which table contains the target information.

```python
for filename in [f for f in os.listdir(DIR) if f.endswith('.zip')]:
    df = pd.read_csv(os.path.join(DIR, filename), nrows=NROWS)
    if 'TARGET' in df.columns:
        display('FILENAME: {}'.format(filename))
        display(df[['SK_ID_CURR','TARGET']].head())
```
